```python
import os
import requests
import urllib.parse
import json
from dotenv import load_dotenv
import time
```

```python
# 環境変数からAPIキーを読み込む関数
def load_api_key():
    load_dotenv(dotenv_path="./.env", override=True)
    api_key = os.getenv("tweet_API_Key")
    return api_key

api_key = load_api_key()
```

```python
# ツイートを検索してリストアップする関数
def tweet_search(query,tweet_count):
    url = "https://api.twitterapi.io/twitter/tweet/advanced_search"
    headers = {"X-API-Key": api_key}

    querystring = {
        "queryType": "Top",
        "query": query
    }

    all_tweets = []
    while len(all_tweets) < tweet_count+1:
        response = requests.get(url, headers=headers, params=querystring)
        data = response.json()
        tweets = data.get("tweets", [])
        all_tweets.extend(tweets)
        # 次ページがなければ終了
        if not data.get("has_next_page"):
            break
        # next_cursor をセットして次ページ取得
        querystring["cursor"] = data["next_cursor"]
        # API制限回避
        time.sleep(7)
    print(f"最終取得件数: {len(all_tweets)}")
    return all_tweets

query = "乃木坂46"
tweet_cnt = 10
tweets = tweet_search(query, tweet_cnt)
```

```python
# ツイートをJSONファイルに保存する関数
def save_tweets_to_json(tweets):
    with open("nogizaka_tweets.json","w",encoding="utf-8") as f:
        json.dump(tweets,f,ensure_ascii=False,indent=2)
```

```python
# Twitter APIをテストで呼び出す関数
def call_tweet_api():
    userName = os.getenv("userName")
    url = 'https://api.twitterapi.io/twitter/user/followings?userName=' + userName
    headers = {'x-api-key': api_key}
    response = requests.get(url, headers=headers)
    return response

response = call_tweet_api()
print(response.json())
```

```python
# Twitter APIを使ってツイートを検索する関数
def search_tweets(query):
    url = "https://api.twitterapi.io/twitter/tweet/advanced_search"
    querystring = {"queryType":"Top","query": query}
    headers = {"X-API-Key": api_key}
    response = requests.get(url, headers=headers, params=querystring)
    return response.json()

query = "乃木坂46"
search_tweets(query)
```
